# Custo medio Ponderado de capital
### CMPC ou WACC

ATIVO | PASSIVO
        Custo de terceiros (credores)
      -------------------------------  + Somados tem que dar 100%
        Custo Proprio


Capital Prorpio =  60 m (P.L)
Capital de Terceiros = 40 m 

Total = 100 m

(pode ser valores ou percentual 'tanto faz')

Custo de capital de terceiros
kd = 12% a.a 

Custo de capital proprio
Ks = 18% a.a

WACC ou CMPC = (DIVIDAS (capital de terceiros) * KD) + (P.l * Ks)
               ------------------------------------------------
                            DIVIDAS  + P.L

```python
import numpy as np

#Exemplo

def calcularKd(ka, taxa):
    return (ka * (1 - taxa))

def calcularKs(dadosKs):
    return dadosKs['txLvRisco'] + (dadosKs['beta'] * (dadosKs['txRisco'] - dadosKs['txLvRisco']))

def calculaKsComPremio(dadosKs):
    return dadosKs['txLvRisco'] + (dadosKs['beta'] * (dadosKs['premioDeRisco']))
    

def calculaWacc(array_parametros):
    dados = array_parametros;
    return ((dados['capitalProprio'] * dados['ks']) + (dados['capitalTerceiro'] * dados['kd'])) / 100

dadosWacc = {
    "capitalTerceiro": 40,
    "kd": calcularKd(8.8255, 0.34),
    "capitalProprio": 60 - 40,
    "ks": 0.12
}

WACC = calculaWacc(dadosWacc)

# Porcentagem
print WACC
```
